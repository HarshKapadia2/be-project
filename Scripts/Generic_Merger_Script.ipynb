```python
import pandas as pd
import numpy as np
```

```python
dataset_bowlers = pd.read_excel('../dataset_2019.xlsx','Bowlers')
dataset_batsmen = pd.read_excel('../dataset_2019.xlsx','Batsmen')
dataset_auction = pd.read_excel('../dataset_2019.xlsx','Auction')
```

```python
auction_names=dataset_auction.iloc[:,0].values
auction_values=dataset_auction.iloc[:,1].values

batsmen_values=dataset_batsmen.iloc[:,:].values
bowlers_values=dataset_bowlers.iloc[:,:].values
```

```python
dataset_bowlers
```

```python
auction_names=[i.split('[')[0].lower() for i in auction_names]
```

```python
prices_dict={auction_names[i]:auction_values[i] for i in range(len(auction_names))}
prices_dict
```

```python
final_list=[]
for i in range(len(batsmen_values)):
    temp=batsmen_values[i]
    name_key=batsmen_values[i][0].lower()
    auction_price=prices_dict.get(name_key,0)
    temp=np.append(temp,auction_price)
    final_list.append(temp)
final_list=np.array(final_list)
final_list
```

```python
batsmen_final_list=pd.DataFrame(final_list,columns=['Player','Inns','NO','Runs','Avg','BF','SR','Price'])
```

```python
final_list=[]
for i in range(len(bowlers_values)):
    temp=bowlers_values[i]
    name_key=bowlers_values[i][0].lower()
    auction_price=prices_dict.get(name_key,0)
    temp=np.append(temp,auction_price)
    final_list.append(temp)
final_list=np.array(final_list)
final_list
```

```python
bowlers_final_list=pd.DataFrame(final_list,columns=['Player','Ov','Wkts','Avg','Econ','SR','Price'])
```

```python
import xlsxwriter
writer = pd.ExcelWriter('../final_dataset_2019_v3.xlsx', engine='xlsxwriter')
batsmen_final_list.to_excel(writer, sheet_name='Batsmen',index=False)
bowlers_final_list.to_excel(writer, sheet_name='Bowlers',index=False)
writer.save()
```
