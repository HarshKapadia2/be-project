```python
import numpy as np
import pandas as pd
```

```python
batsmen_df=pd.read_excel('final_dataset_2019_v2.xlsx','Bowlers')
```

```python
X = batsmen_df.iloc[:, 1:-1].values
y = batsmen_df.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, :])
X[:, :] = imputer.transform(X[:, :])
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# X_test
# [9.75e+02, 6.82e+02, 2.44e+01, 1.43e+02]
# regressor.predict([[9.75e+02, 6.82e+02, 2.44e+01, 1.43e+02]])
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print('Coefficients: \n', regressor.coef_)
# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred))
```
